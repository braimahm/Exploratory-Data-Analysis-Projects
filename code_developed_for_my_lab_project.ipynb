```python
 ''' The code calculates rolling average
 and derivative of a 10nm reflectance'''

 #importing packages
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
#loading csv data into my code
sk_1_grass = pd.read_csv("/052423_SK_1_Grass.csv" )
sk_1_grass
```

```python
#this code drops the wavelength column and then calculates the mean of each row of reflectance
#we name the new column Mean
sk_1_grass['Mean']=sk_1_grass.drop('Wavelength', axis=1).mean(axis=1)

#we multiply the average of the reflectance by 100 to get relectance in percetange
sk_1_grass['Mean'] = sk_1_grass['Mean']*100
sk_1_grass['Mean']


```

```python
'''We want to
    plot the average reflectance for 1nm wavlength
'''
plt.plot(sk_1_grass.Wavelength, sk_1_grass.Mean)
plt.title("1nm Average Reflectance")
plt.xlabel("Wavelength")
plt.ylabel("Reflectance")
```

```python

'''we want to calculate the moving average of the reflectance'''

# we calculate the simple moving average, using window size of 10nm

sk_1_grass['moving_average'] = sk_1_grass['Mean'].rolling(10).mean()

# we want to drop the NA values for the first nine rows after calculating the moving average
#sk_1_grass['moving_average'].dropna(inplace=True)

# we make sure the columns are indicated in the table
sk_1_grass.columns
```

```python
sk_1_grass.moving_average[0:30] # we are just looking at the moving average
```

```python
'''What we want to do is to match the 10nm dataframe created above with the moving
average reflectance. The first moving average starts from the 10th row (row 9).
We want to select the moving averages in intervals of 10 and macth
them with the 10nm wavelengths with centre
starting from 355nm.'''

#We slice the moving averages 10nm intervals


len(sk_1_grass.moving_average[9:1922:10])



```

```python
#What we want to do next is to match the centre of the moving averages to the centre of the 10nm bands
#we start from 355 to 365, etc. in 10nm intervals
ten_nm = [i for i in range(355,2275,10)]
len(ten_nm)




```

```python
# we convert the list  of 10nm centres to dataframe

W_ten_nm = pd.DataFrame({"ten_nm": ten_nm, "centered_mov_av":sk_1_grass.moving_average[9:1922:10]} )
W_ten_nm
```

```python
'''We want to
    plot the moving averages against the 10nm centers
'''
plt.plot(W_ten_nm.ten_nm, W_ten_nm.centered_mov_av)
plt.title("10nm Moving Average Reflectance")
plt.xlabel("Wavelength")
plt.ylabel("Reflectance")
```

```python
''' We want to perform the first derivative of the 10 nm spectra drdl

             '''
wavelegth_diff = W_ten_nm.ten_nm.diff(periods=2)
reflectance_diff = W_ten_nm.centered_mov_av.diff(periods=2)
derivative = reflectance_diff/wavelegth_diff
derivative
```

```python
""" we create a new 10nm list starting from 365nm"""
'''we drop the first endpoint (350nm) and move
the 10nm wavlength one row up to match the centers ,
                             '''
new_ten_nm = [i for i in range(365,2285,10)]
len(new_ten_nm)
```

```python


len (derivative)
```

```python
derivative_reflectance = pd.DataFrame({"new_ten_nm": new_ten_nm, "derivative": derivative})
derivative_reflectance
```

```python
'''We want to
    plot the derivative spectra against the 10nm centers
    starting from 365nm.
    What we did no do here is that we eliminated the first end point (350nm)
    We also kept the last endpoint to 20nm difference but not 10nm
'''
plt.plot(derivative_reflectance.new_ten_nm, derivative_reflectance.derivative)
plt.title("10nm derivative spectra")
plt.xlabel("Center Wavelength")
plt.ylabel("Derivative Reflectace")
```
